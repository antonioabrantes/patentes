```python
# Lista de títulos
data_inicio = '2007-01-01'
data_fim = '2007-12-31'
base_url = 'https://api.patentsview.org/patents/query?q={%22_and%22:[{%22_gte%22:{%22patent_date%22:%22'
url = base_url + f'{data_inicio}' + '%22}},{%22_lte%22:{%22patent_date%22:%22' + f'{data_fim}' + '%22}}]}'
url = url + '&o={"page":1,"per_page":50}'
print(url)
r=requests.get( url )
y = json.loads(r.text)
for n in range(50):
    print(y["patents"][n]["patent_title"])
```

```python
# Lista de inventores
inventor_last_name = 'Whitney'
data = '2007-01-04'
base_url = 'https://api.patentsview.org/patents/query?q={%22_and%22:[{%22_gte%22:{%22patent_date%22:%22'
url = base_url + f'{data}' + '%22}},{%22inventor_last_name%22:%22'
url = url + f'{inventor_last_name}' + '%22}]}&f=[%22patent_number%22,%22patent_date%22,%22inventor_last_name%22]'
url = url + '&o={"page":1,"per_page":50}'
print(url)
r=requests.get( url )
y = json.loads(r.text)

for n in range(50):
    print(y["patents"][n]["patent_number"])
    z = len(y["patents"][n]["inventors"])
    for k in range(z):
        print(y["patents"][n]["inventors"][k]["inventor_last_name"])
    print()
```

```python
# Contagem de CPC
# https://patentsview.org/apis/api-endpoints/cpc
cpc = 'G12'
base_url = 'https://api.patentsview.org/patents/query?q={"cpc_subsection_id":"'
url = base_url + f'{cpc}' + '"}&f=["cpc_subsection_id","cpc_subsection_title","cpc_total_num_patents"]'
url = url + '&o={"page":1,"per_page":50}'
print(url)
r=requests.get( url )
y = json.loads(r.text)
print(len(y["patents"][0]["cpcs"]))
z = len(y["patents"][0]["cpcs"])
for n in range(z):
    if y["patents"][0]["cpcs"][n]["cpc_subsection_id"]==cpc:
        print(y["patents"][0]["cpcs"][n]["cpc_subsection_id"])
        print(y["patents"][0]["cpcs"][n]["cpc_subsection_title"])
        print(y["patents"][0]["cpcs"][n]["cpc_total_num_patents"])
```

```python
# Contagem de CPC
# https://patentsview.org/apis/api-endpoints/cpc
cpc = 'G12'
base_url = 'https://api.patentsview.org/patents/query?q={"cpc_subsection_id":"'
url = base_url + f'{cpc}' + '"}&f=["cpc_subsection_id","cpc_subsection_title","patent_number"]'
url = url + '&o={"page":1,"per_page":50}'
print(url)
r=requests.get( url )
y = json.loads(r.text)
print(len(y["patents"]))
z = len(y["patents"])
for n in range(z):
    print("US",y["patents"][n]["patent_number"].strip())
    w = len(y["patents"][n]["cpcs"])
    for m in range(w):
        print(y["patents"][n]["cpcs"][m]["cpc_subsection_id"])
        print(y["patents"][n]["cpcs"][m]["cpc_subsection_title"])
        print("")
```
